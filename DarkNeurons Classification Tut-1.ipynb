```python
from DarkNeurons import Classify_Images
```

```python
directory = 'C:/Users/TusharGoel/Desktop'
training_directory = r'C:\Users\TusharGoel\Desktop\flowers'
```

```python
cnn = Classify_Images(working_directory = directory,
                     output_directory = directory)
```

```python
# Preprocessing the Data
train,val,labels = cnn.Preprocess_the_Image(method='directory',num_classes = 5,training_image_directory = training_directory,
                                       model_name = 'InceptionV3',train = True)
```

```python
model = cnn.Create_the_Model()
```

```python
metrics = ['accuracy']
model = cnn.Train_the_Model(model=model,
                           rebuild = True,
                           epochs = 5,
                            train_data_object = train,
                           loss = 'categorical_crossentropy',
                           metrics = metrics)
```

```python
model.save('Flowers_Model.h5')
```

```python
test = cnn.Preprocess_the_Image(method = 'directory',train =False,user_model=model,
                           test_image_directory = r'C:\Users\TusharGoel\Desktop\test_image_directory',
                           num_classes = 5)
```

```python
cnn.Predict_from_the_Model(labels = labels,model=model,generator = test,top = 5)
```
